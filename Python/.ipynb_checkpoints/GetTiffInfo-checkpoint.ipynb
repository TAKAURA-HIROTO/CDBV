```python
dis = geodesic((34.6572940, 135.5918912), (34.6572940, 135.5980244)).m
picinch = 0.1662453930192411
pix = dis / picinch
pix
```

```python
#左上から反時計回り
point0002 = [(135.5918912,  34.6572940), (135.5918912,  34.6529279), (135.5980244,  34.6529279), (135.5980244,  34.6572940)]
point0003 = [( 135.5862067,  34.6577636), ( 135.5862067,  34.6533974), ( 135.5923323,  34.6533974), ( 135.5923323,  34.6577636)]
point0004 = [( 135.5804335,  34.6582547), ( 135.5804335,  34.6538886), ( 135.5865668,  34.6538886), ( 135.5865668,  34.6582547)]
point0005 = [( 135.5753427,  34.6590715), ( 135.5753427,  34.6547016), ( 135.5814682,  34.6547016), ( 135.5814682,  34.6590715)]
```

```python
#検出範囲のポリゴンデータ作成
ring0002 = ogr.Geometry(ogr.wkbLinearRing)
ring0003 = ogr.Geometry(ogr.wkbLinearRing)
ring0004 = ogr.Geometry(ogr.wkbLinearRing)
ring0005 = ogr.Geometry(ogr.wkbLinearRing)
for i in range(4):
    ring0002.AddPoint(point0002[i][0], point0002[i][1])
    ring0003.AddPoint(point0003[i][0], point0003[i][1])
    ring0004.AddPoint(point0004[i][0], point0004[i][1])
    ring0005.AddPoint(point0005[i][0], point0005[i][1])
    
poly0002 = ogr.Geometry(ogr.wkbPolygon)
poly0002.AddGeometry(ring0002)
poly0003 = ogr.Geometry(ogr.wkbPolygon)
poly0003.AddGeometry(ring0003)
poly0004 = ogr.Geometry(ogr.wkbPolygon)
poly0004.AddGeometry(ring0004)
poly0005 = ogr.Geometry(ogr.wkbPolygon)
poly0005.AddGeometry(ring0005)

detectionrange = {'WKT': [poly0002.ExportToWkt(), poly0003.ExportToWkt(), poly0004.ExportToWkt(), poly0005.ExportToWkt()]}
detectionrange
```

```python
detectionrange = pd.DataFrame(detectionrange)
```
